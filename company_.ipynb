```python
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.linear_model import LinearRegression

company = pd.read_csv("/content/company_.csv")

x = company['Year']
x_new = x.values.reshape(-1, 1)
y = company['No of Employee']

x_train, x_test, y_train, y_test = train_test_split(x_new, y, random_state=42, test_size=0.2)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(f"Predicted : {y_pred}")

# Evaluation metrics for regression
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"R-squared : {r2}")
print(f"Mean Squared Error : {mse}")

# Plotting the results
fig = plt.figure()
fig.patch.set_facecolor('xkcd:purple')
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('Year')
plt.ylabel('No of Employee',)
plt.title('Number of employees in a company changes over time, taking into account different years\n')
plt.legend()
plt.show()
```
